# Exerc√≠cios de Match/Case

---

**Exerc√≠cio 1:** Tratamento de Erros (Web Scraping)
Na an√°lise de dados, frequentemente raspamos dados da internet. Os servidores retornam c√≥digos (200 = OK, 404 = N√£o encontrado, 500 = Erro no servidor).

- Crie uma vari√°vel codigo_erro = 404.
- Use match codigo_erro:
- case 200: imprima "Sucesso: Dados coletados".
- case 404: imprima "Erro: P√°gina n√£o encontrada".
- case 500: imprima "Erro: Servidor indispon√≠vel".
- case _: (o caso padr√£o): imprima "Erro desconhecido".

---

**Exerc√≠cio 2:** Categoriza√ß√£o de Clientes (Com Guard Clauses)
Vamos refazer uma l√≥gica parecida com a da m√©dia escolar, mas para Score de Cr√©dito (semelhante ao exerc√≠cio anterior de if, mas usando a sintaxe nova).

- Crie score = 800.
- Use match score:
- case score if score < 300: imprima "Recusado".
- case score if 300 <= score < 700: imprima "Em an√°lise".
- case score if score >= 700: imprima "Aprovado com Cr√©dito".

`Obs: Note que no Exerc√≠cio 2, como estamos comparando n√∫meros (maior/menor), precisamos repetir o nome da vari√°vel (case score if score < 300). No Exerc√≠cio 1, como √© o valor exato, basta colocar o n√∫mero (case 200).`

```python
# EXERCIC√çO 1

codigo_erro = 300 # Atribui√ß√£o da vari√°vel

match codigo_erro:  # Cria√ß√£o do comando match
  case 200: # Primeiro teste l√≥gico do case
    print("Sucesso: Dados coletados")
  case 404: # Segundo teste l√≥gico do case
    print("Erro: P√°gina n√£o encontrada")
  case 500: # Terceiro teste l√≥gico do cas
    print("Erro: Servidor indispon√≠vel")
  case _: # Ultimo teste l√≥gico do case caso nenhuma outro teste l√≥gico tenha sido efetuado
    print("Erro desconhecico")
```

```python
# EXERCIC√çO 2

score = 80 # Atribui√ß√£o da vari√°vel

match score:  # Cria√ß√£o do comando match
  case score if score < 300:  # Primeiro teste l√≥gico, verificando se o score √© menor que 300
    print("Recusado")
  case score if 300 <= score < 700:  # Segundo teste l√≥gico, verificando se 300 √© menor ou igual ao score e score √© menor que 700
    print("Em an√°lise")
  case score if score >= 700:  # Terceiro teste l√≥gico, verificando se o score √© maior ou igual que 700
    print("Aprovado com Cr√©dito")
```

# IF e WHILE

Vamos fazer dois exerc√≠cios focados nessas diferen√ßas.

**Exerc√≠cio 1**: A Lista de Pre√ßos (for)
Voc√™ tem uma lista de pre√ßos e quer aplicar um desconto de 10% em todos eles para ver como ficaria.

- Crie uma lista: precos = [100, 200, 500]

- Crie um for que passe por cada preco na lista.

- Dentro do loop, imprima o pre√ßo com desconto (multiplique por 0.9).

---

**Exerc√≠cio 2**: A Meta de Vendas (while)
Voc√™ quer simular quanto tempo leva para bater uma meta acumulada.

- Crie uma vari√°vel total_vendas = 0.

- Crie um loop while que rode enquanto total_vendas for menor que 1000.

- Dentro do loop:

- Adicione um valor fixo (ex: total_vendas = total_vendas + 100).

- Imprima o valor atual: print("Venda processada. Total:", total_vendas).

```python
# EXERCICIO 1

precos = [100, 200, 500] # Cria√ß√£o da V√°riavel

for i in precos: # Inicio do 'for', indicando que 'i' seria a representa√ß√£o de cada item
  print("Com descontos os pre√ßos ficam", i * 0.9) # print com texto e multiplica√ß√£o por 'i', representando que cada 'i'(item) ser√° multiplicado por 0.9 e ter√° como saida um texto
```

```python
# EXERCICIO 2

total_vendas = 0 # Cria√ß√£o da V√°riavel

while total_vendas < 1000: # Inicio do While, perguntando "enquanto total_vendas for menor que 1000"
  total_vendas = total_vendas + 100 # Resposta do While, total_vendas ser√° igu√°l a total_vendas + 100
  print("Venda processada. Total:", total_vendas) # Print representado para quando o While finalizar, quando teste anterior chegar a 1000, o while se encerra e o print anunciando que o total chegou a igual ou ultrapassou 1000
```
